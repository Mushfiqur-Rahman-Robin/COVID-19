```python
import numpy as np
import pandas as pd
import folium
```

```python
df= pd.read_csv('C:/Users/USER/Desktop/covid19_apr19.csv')
df.head()
```

```python
df.shape
```

```python
bd_lat = 23.6850
bd_long = 90.3563
bd_map = folium.Map(location=[bd_lat, bd_long], zoom_start=8)

for Latitude, Longitude, label in zip(df.Latitude, df.Longitude, df.I_number):
    folium.CircleMarker(
        [Latitude, Longitude],
        radius=np.log(label*100),  #define how big you want the circle markers to be
        color='black',
        fill=True,
        popup=str(label),
        fill_color='red',
        fill_opacity=0.8
    ).add_to(bd_map)

#show map
bd_map
```
